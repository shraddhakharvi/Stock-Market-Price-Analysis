```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from alpha_vantage.timeseries import TimeSeries
```

```python
data = pd.read_csv("AAPL_stock_data.csv")
```

```python
data.head()
```

```python
# Check for missing values
print("Missing values:\n", data.isnull().sum())

# Drop missing values if any
data.dropna(inplace=True)

# Add additional columns (Daily Change & Percentage Change)
data["Daily Change"] = data["Close"] - data["Open"]
data["Percentage Change"] = (data["Daily Change"] / data["Open"]) * 100

# Display the cleaned data
print(data.head())
```

```python
data.head()
```

```python
symbol = "AAPL"  # Replace with the stock symbol you are analyzing
```

# Stock Closing Price Over Time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x=data.index, y="Close", color="blue", label="Closing Price")
plt.title(f"{symbol} Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.show()
```

#  Stock Price Distribution

```python
plt.figure(figsize=(10, 5))
sns.histplot(data["Close"], bins=50, kde=True, color="purple")
plt.title(f"Distribution of {symbol} Closing Prices")
plt.xlabel("Stock Price (USD)")
plt.show()
```

#  Daily Percentage Change Over Time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x=data.index, y="Percentage Change", color="red")
plt.axhline(y=0, color="black", linestyle="--")
plt.title(f"{symbol} Daily Percentage Change Over Time")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.show()
```

# Stock Trading Volume Over Time

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x=data.index, y="Volume", color="green")
plt.title(f"{symbol} Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
# Save the stock data to a CSV file
filename = f"{symbol}_stock_data.csv"
data.to_csv(filename)

print(f"Data saved successfully as {filename}")
```
